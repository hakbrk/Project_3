```python
import pandas as pd
```

```python
Rigs= pd.read_csv("/Users/shilpa/Desktop/Desktop/RICE_DATA/Project_3/Shilpa_Project_3/Organised folder/Yearly_Average copy.csv")
Rigs.head(3)
```

```python
land_rigs = Rigs[Rigs['Type']=='L'].drop('Type', axis = 1)
land_rigs.set_index(['Unnamed: 0'], inplace=True)
headers = land_rigs.columns.tolist()
df_list = []
for column in headers:
    df=land_rigs[[column]]
    df['year']=column
    df.rename(columns={column:'Count'}, inplace=True)
    df_list.append(df)
new_land=pd.concat(df_list)
new_land.reset_index(drop=False, inplace=True)
new_land['Type']="In-Land"
new_land.rename(columns={'Unnamed: 0':'country'},inplace=True)
new_land
```

```python
offshore_rigs = Rigs[Rigs['Type']=='O'].drop('Type', axis = 1)
offshore_rigs.set_index(['Unnamed: 0'], inplace=True)
headers = offshore_rigs.columns.tolist()
df_list = []
for columns in headers:
    df = offshore_rigs[[column]]
    df['year'] = column
    df.rename(columns={column:'Count'}, inplace=True)
    df_list.append(df)
new_shore = pd.concat(df_list)
new_shore.reset_index(drop = False, inplace = True)
new_shore['Type'] = 'Off-Shore'
new_shore.rename(columns={'Unnamed: 0':'country'},inplace=True)
new_shore
    
```

```python
combine_df=pd.concat([new_land, new_shore])
combine_df
```

```python
combine_df.to_csv('Yearly_Rigs_data.csv',index=False,header=True)
```
